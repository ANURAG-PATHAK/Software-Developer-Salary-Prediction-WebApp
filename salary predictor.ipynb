```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data/survey_results_public.csv")
df.head()
```

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
df = df.rename({"ConvertedCompYearly" : "Salary"}, axis = 1)
df
```

```python
df = df[df["Salary"].notnull()]
df
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df = df[df["Employment"] == "Employed, full-time"]
df = df.drop("Employment",axis=1)
```

```python
df
```

```python
df["Country"].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = "Other"
    return categorical_map
```

```python
country_map = shorten_categories(df["Country"].value_counts(), 400)
df["Country"] = df["Country"].map(country_map)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,7))
df.boxplot("Salary", "Country",ax=ax)
plt.suptitle("Salary (US$) v Country")
plt.title("")
plt.ylabel("Salary")
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 120000]
df = df[df["Salary"] >= 5000]
df = df[df["Country"] != "Other"]
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,7))
df.boxplot("Salary", "Country",ax=ax)
plt.suptitle("Salary (US$) v Country")
plt.title("")
plt.ylabel("Salary")
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x == "Less than 1 year":
        return 0.5
    elif x == "More than 50 years":
        return 50
    else:
        return float(x)
df["YearsCodePro"] = df["YearsCodePro"].apply(clean_experience)
```

```python
df["YearsCodePro"].unique()
```

```python
df["EdLevel"].unique()
```

```python
def clean_education(x):
    if "Bachelor’s degree" in x:
        return "Bachelor’s degree"
    elif "Master’s degree" in x:
        return "Master’s degree"
    elif "Other doctoral" in x or "Professional degree" in x:
        return "Doctorate or Professional degree"
    else:
        return "Less than Bachelor’s degree"
df["EdLevel"] = df["EdLevel"].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df["EdLevel"] = le_education.fit_transform(df["EdLevel"])
df["EdLevel"].unique()
```

```python
le_country = LabelEncoder()
df["Country"] = le_country.fit_transform(df["Country"])
df["Country"].unique()
```

```python
x = df.drop("Salary",axis=1)
x
```

```python
y = df["Salary"]
y
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y.values)
```

```python
y_predicted = lr.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y,y_predicted))
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree_reg = DecisionTreeRegressor(random_state=0)
dtree_reg.fit(x,y.values)
```

```python
y_predicted = dtree_reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y,y_predicted))
error
```

```python
from sklearn.ensemble import RandomForestRegressor
rand_forest_reg = RandomForestRegressor(random_state=0)
rand_forest_reg.fit(x,y.values)
```

```python
y_predicted = rand_forest_reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y,y_predicted))
error
```

```python
from sklearn.model_selection import GridSearchCV
max_depth = [None, 2, 4, 6, 8, 10, 12]
parameters = {"max_depth": max_depth}
regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring="neg_mean_squared_error")
gs.fit(x,y.values)
```

```python
regressor = gs.best_estimator_
regressor.fit(x,y.values)
```

```python
y_predicted = regressor.predict(x)
y_predicted
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y,y_predicted))
error
```
